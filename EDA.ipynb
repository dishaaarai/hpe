```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\disha\\Downloads\\data_1.csv",parse_dates=['START_TIME'])
```

```python
df.set_index('START_TIME', inplace=True)
```

```python
for col in df.columns:
    plt.plot(df.index, df[col])
    plt.title(col)
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.savefig(f"{col}_line_chart.png")
    plt.show()
```
